### 15.2.1 修改标签⽂字和线条粗细

```python
import matplotlib.pyplot as plt
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)
 # 设置图题并给坐标轴加上标签
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
 # 设置刻度标记的样式
ax.tick_params(labelsize=14)
plt.show()
```

### 15.2.2 校正绘图

```python
import matplotlib.pyplot as plt
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)
# 设置图题并给坐标轴加上标签
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
 # 设置刻度标记的样式
ax.tick_params(labelsize=14)
plt.show()
```

### 15.2.3 使⽤内置样式

```python
import matplotlib.pyplot as plt

# 列出所有可用样式
print(plt.style.available)

# 使用其中一个样式
plt.style.use('Solarize_Light2')

# 创建一个简单的图表
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sample Plot with Solarize_Light2 Style')

plt.show()
```
